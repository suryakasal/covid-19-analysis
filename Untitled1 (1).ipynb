```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('covid_vaccine_statewise.csv')
```

```python
df.head()
```

```python
df.rename(columns={'Updated On' : 'VaccineDate'},inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=['Sputnik V (Doses Administered)','AEFI','18-44 Years (Doses Administered)','45-60 Years (Doses Administered)','60+ Years (Doses Administered)'],axis=1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
perc = 75.0 # Here N is 75
min_count =  int(((100-perc)/100)*df.shape[1] + 1)
mod_df = df.dropna( axis=0, 
                    thresh=min_count)
```

```python
mod_df.tail()
```

```python
mod_df.isnull().sum()
```

```python
perc = 80.0 # Here N is 75
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
data = mod_df.dropna( axis=1, 
                    thresh=min_count)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
data=data[data.State!='India']
```

```python
data.isnull().sum()
```

```python
data
```

```python
data=data.fillna(0)
```

```python
data
```

```python
data['VaccineDate'] = pd. to_datetime(data['VaccineDate'],format='%d/%m/%Y')
data['year'] = pd. DatetimeIndex(data['VaccineDate']). year
data['month'] = pd. DatetimeIndex(data['VaccineDate']). month
```

```python
data.head()
```

```python
d={1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
data.month=data.month.map(d)
```

```python
data.head()
```

```python
data.to_excel('vaccination.xlsx', index=False)
```
