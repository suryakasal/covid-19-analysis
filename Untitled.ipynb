```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('covid_19_india.csv')
```

```python
data.head()
```

```python
data=data.drop(columns=['Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

```python
data.head()
```

```python
data['Date'] = pd. to_datetime(data['Date'],format='%Y-%m-%d')
data['year'] = pd. DatetimeIndex(data['Date']). year
data['month'] = pd. DatetimeIndex(data['Date']). month
```

```python
data.head()
```

```python
data.describe()
       
```

```python
print(data.dtypes)
```

```python
d={1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
data.month=data.month.map(d)
```

```python
data.head()
```

```python
data['ActiveCase']=data['Confirmed']-(data['Cured']+data['Deaths'])
```

```python
data.head()
```

```python
data['RecoveryRate']=(data['Cured']/data['Confirmed'])*100
```

```python
data['MortalityRate']=(data['Deaths']/data['Confirmed'])*100
```
